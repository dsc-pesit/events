# Data Cleaning [4:20]

The **Current Ver** column has some values as "Varies with device".

We want to convert all of these values to NaN

```python
#replacing values 
df['Current Ver'] = df['Current Ver'].replace(
    'Varies with device',np.nan)
df
```

```python
# drop all NaNs
#df = df.dropna(subset=('Current Ver',))
df.dropna().size
```
