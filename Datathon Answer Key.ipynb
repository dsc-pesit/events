## Problem 3

Which category of products gives the max amount of discounts?

**Check your answer**:

Category: Household

Discount: 2836.0

**Solution**

```python
import numpy as np 

df['discount'] =  pd.to_numeric(df['discount'],errors='coerce')

# group by category and apply max() to get max of each category 
grouped = df[['category','discount']].groupby(['category']).max()

grouped
```

```python
#to get the row with highest discount

grouped[ grouped['discount'] == grouped['discount'].max()]
```

## Problem 4

Compare the average difference between MRP and SP for products of each category.

**Check your answer** 

The answer for this is the entire column. Hence just compare your answer with the table below.

**Solution**

```python
# force convert to numeric types - eliminates all rows with non numeric values
df['MRP'] =  pd.to_numeric(df['MRP'],errors='coerce')
df['SP'] =  pd.to_numeric(df['SP'],errors='coerce')

# make a new column called difference  
df['difference'] = df['MRP'] - df['SP']

# apply group by and use mean()
grouped = df[['category','difference']].groupby(['category']).mean()

grouped
```
